```python
# 답안을 한줄로 찍는것과 무작위로 찍는것의 결과를 비교하는 시뮬레이션

import random
import numpy as np
import matplotlib.pyplot as plt

label = [(i%5)+1 for i in range(100)]  # 1~5가 각 인덱스별로 균일하게 있는 라벨

history = [0]*101  # 매 시행마다 정답 갯수에 해당하는 인덱스를 +1
results = []  # 정답을 저장
mean_history = []  # 지금까지 나온 정답 갯수의 평균을 저장
```

```python
# 시행

for trial in range(100000):
    answer = [random.randint(1, 5) for i in range(100)]
    #print(answer)

    # 정답이면 1, 오답이면 0
    correct = [1 if label[i] == answer[i] else 0 for i in range(100)]
    #print(correct)

    results.append(correct.count(1))
    history[correct.count(1)] += 1
    mean_history.append(sum(results)/len(results))
```

```python
print("min score :", np.unique(results)[0])
print("max score :", np.unique(results)[-1])

plt.title(f"trial {trial+1} mean distribution")
plt.plot(history)
plt.show()

plt.title(f" trial {trial+1} total mean")
plt.plot(mean_history)
plt.show()
```
